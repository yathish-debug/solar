```python
import pandas as pd
import requests
file = requests.get("https://weather.com/en-IN/weather/tenday/l/INKA0344:1:IN")
print(file)          
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(file.content, 'html.parser')
```

```python
print(soup)
```

```python
list=[]
```

```python
l=[]
day = soup.findAll("h3",{"class":"DetailsSummary--daypartName--2FBp2"})
hightemp = soup.findAll("span",{"class":"DetailsSummary--highTempValue--3Oteu"})
lowtemp = soup.findAll("span",{"class":"DetailsSummary--lowTempValue--3H-7I"})
data = soup.findAll("span",{"class":"DetailsSummary--extendedData--365A_"})

from csv import DictWriter
field_names = ['day','hightemp','lowtemp',
               'data']
for i in range(0 ,10):
    dict={'day':day[i].text,'hightemp':hightemp[i].text,'lowtemp':lowtemp[i].text,'data':data[i].text}
    l.append(dict)

import pandas as pd
convert = pd.DataFrame(l)
convert.set_index('day',inplace=True)
convert.to_csv("finaldata.csv")
print(convert)


  


```
